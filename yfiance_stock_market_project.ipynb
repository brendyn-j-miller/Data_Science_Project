```
import yfinance as yf

msft = yf.Ticker("MSFT")

# get stock info
print(msft.info)

# get historical market data
hist = msft.history(period="1y")
```

```
import matplotlib.pyplot as plt
import seaborn
# Plot everything by leveraging the matplotlib package
hist['Close'].plot(figsize=(16, 9))
plt.title('Closing Prices for MSFT for the past year')
plt.ylabel('Closing Price of Stock per Share ($)')
```

```
# Download stock data and then export as CSV
data_df = yf.download("AAPL", start = "2020-02-01", end = "2020-03-20")
data_df.to_csv('aapl.csv')
```
